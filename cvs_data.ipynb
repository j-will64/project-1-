```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
treasure_bills_df = pd.read_excel(r"data\treasuary bills .xlsm")

# Print data
treasure_bills_df.head()
```

```python
# 4 week bank discount
four_bd_mean = treasure_bills_df["4 WEEKS BANK DISCOUNT"].mean()
four_bd_median = treasure_bills_df["4 WEEKS BANK DISCOUNT"].median()
four_bd_min = treasure_bills_df["4 WEEKS BANK DISCOUNT"].min()
four_bd_max = treasure_bills_df["4 WEEKS BANK DISCOUNT"].max()
four_bd_var = treasure_bills_df["4 WEEKS BANK DISCOUNT"].var()

four_bd_data = pd.DataFrame({"4 Week BD mean": [four_bd_mean],
                             "4 Week BD mediam": [four_bd_median],
                             "4 Week BD min": [four_bd_min],
                             "4 Week BD max": [four_bd_max],
                             "4 Week BD var": [four_bd_var]})

four_bd_data
```

```python
# 4 week coupon equivalent
four_ce_mean = treasure_bills_df["4 WEEKS COUPON EQUIVALENT"].mean()
four_ce_median = treasure_bills_df["4 WEEKS COUPON EQUIVALENT"].median()
four_ce_min = treasure_bills_df["4 WEEKS COUPON EQUIVALENT"].min()
four_ce_max = treasure_bills_df["4 WEEKS COUPON EQUIVALENT"].max()
four_ce_var = treasure_bills_df["4 WEEKS BANK DISCOUNT"].var()

four_ce_data = pd.DataFrame({"4 Week CE mean": [four_ce_mean],
                             "4 Week CE mediam": [four_ce_median],
                             "4 Week CE min": [four_ce_min],
                             "4 Week CE max": [four_ce_max],
                             "4 Week BD var": [four_ce_var]})

four_ce_data
```
