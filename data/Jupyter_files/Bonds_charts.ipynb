```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
treasure_bills_df = pd.read_excel(r"..\Excel_files\treasuary bills .xlsm")

# Print a csv file
treasure_bills_df.to_csv("../CSV_files/treasure_bills.csv", index=False)


# Print data
treasure_bills_df.head()
```

```python
# Assuming "Date" is in datetime format. If not, convert it to datetime.
treasure_bills_df['Date'] = pd.to_datetime(treasure_bills_df['Date'])

# Define the columns for each line on the graph
columns = ["4 WEEKS BANK DISCOUNT", "8 WEEKS BANK DISCOUNT", "13 WEEKS BANK DISCOUNT", "17 WEEKS BANK DISCOUNT", "26 WEEKS BANK DISCOUNT", "52 WEEKS BANK DISCOUNT"]

# Set the figure size (adjust the width and height as needed)
fig, ax = plt.subplots(figsize=(10, 6))  # Width: 10 inches, Height: 6 inches
 
# Define colors for each line
line_colors = ['#ADD8E6', '#87CEEB', '#B0E0E6', '#0000CD', '#000080', '#00008B']

# Create a line graph for the specified columns with different transparencies
transparencies = [1.0, 0.8, 0.6, 0.4, 0.3, 0.2]  # Adjust transparency values as needed

for i, column in enumerate(columns):
    plt.plot(treasure_bills_df['Date'], treasure_bills_df[column], label=column, color=line_colors[i], alpha=transparencies[i])

# Customize the graph's appearance
plt.title('Treasure Bills Interest Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Interest Rate')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(loc='lower right')  # Display a legend for the lines

# Display the graph
plt.tight_layout()  # Ensure the labels fit within the graph area
plt.savefig("../Charts/Bank_Discount_chart.png")

# Display the graph
plt.show()
```
